##### 2. Pasing Date

```python
# Parse and fix CustomerDOB; handling future date
def parse_customer_dob(dob_str):
    try:
        dt = pd.to_datetime(dob_str, dayfirst=True, errors='coerce')
        # Fix future DOB (e.g., 2094 -> 1994)
        if pd.notnull(dt) and dt > pd.Timestamp.now():
            dt = dt.replace(year=dt.year - 100)
        return dt
    except:
        return pd.NaT

# Parse and fix TransactionDate
def parse_transaction_date(date_str):
    try:
        dt = pd.to_datetime(date_str, dayfirst=True, errors='coerce')
        # Convert 16 -> 2016
        if pd.notnull(dt) and dt.year < 2000:
            dt = dt.replace(year=dt.year + 2000)
        return dt
    except:
        return pd.NaT

# Apply transformations ---
data['CustomerDOB'] = data['CustomerDOB'].apply(parse_customer_dob)
data['TransactionDate'] = data['TransactionDate'].apply(parse_transaction_date)

# --- Ensure dtype is datetime64
data['CustomerDOB'] = pd.to_datetime(data['CustomerDOB'], errors='coerce')
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'], errors='coerce')

# Verify the datatypes
print(data[['CustomerDOB', 'TransactionDate']].dtypes)
print(data[['CustomerDOB', 'TransactionDate']].head(20))
```

##### 3. Combine Transaction date and Time for accurate analysis, such as hourly patterns.

```python
from datetime import datetime, time

# Parse TransactionTime
def parse_transaction_time(t):
    """Convert HHMMSS numeric or string to a proper time object, handling invalid values safely."""
    if pd.isna(t):  # handle NaN
        return None
    t = ''.join(filter(str.isdigit, str(t)))  # remove any non-digit characters
    if len(t) != 6:  # if it's not exactly 6 digits, return None
        return None
    hour = int(t[0:2])
    minute = int(t[2:4])
    second = int(t[4:6])
    return time(hour, minute, second)

# Convert TransactionTime to time objects
data['TransactionTime'] = data['TransactionTime'].apply(parse_transaction_time)

# Combine TransactionDate and TransactionTime
data['TransactionDateTime'] = data.apply(
    lambda row: datetime.combine(row['TransactionDate'].date(), row['TransactionTime'])
                if pd.notnull(row['TransactionDate']) and row['TransactionTime'] is not None else pd.NaT,
    axis=1
)

# Convert to datetime64 (ensures correct dtype)
data['TransactionDateTime'] = pd.to_datetime(data['TransactionDateTime'], errors='coerce')

# Print the three columns ---
print(data[['TransactionDate', 'TransactionTime', 'TransactionDateTime']].head(20))
print(data[['TransactionDate', 'TransactionTime', 'TransactionDateTime']].dtypes)
```

#### 5. Outlier detection

##### Using IQR

```python
# Define a function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"--- {column} ---")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"Number of outliers in {column}: {len(outliers)}\n")
    return outliers

# Detect outliers in TransactionAmount and CustAccountBalance
outliers_transaction = detect_outliers_iqr(data, 'TransactionAmount')
outliers_balance = detect_outliers_iqr(data, 'CustAccountBalance')
```

##### Using Percentile

```python
def detect_outliers_percentile(df, column, lower_pct=1, upper_pct=99):
    lower_bound = np.percentile(df[column], lower_pct)
    upper_bound = np.percentile(df[column], upper_pct)

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"--- {column} ---")
    print(f"{lower_pct}th Percentile: {lower_bound:.2f}")
    print(f"{upper_pct}th Percentile: {upper_bound:.2f}")
    print(f"Number of outliers in {column}: {len(outliers)}\n")
    return outliers

# Detect outliers using percentile method
outliers_transaction = detect_outliers_percentile(data, 'TransactionAmount', lower_pct=1, upper_pct=99)
outliers_balance = detect_outliers_percentile(data, 'CustAccountBalance', lower_pct=1, upper_pct=99)
```

#### Monthly Transaction

```python

# ensure date column exists and is datetime
date_col = 'TransactionDate' if 'TransactionDate' in data.columns else 'TransactionDateTime'
data[date_col] = pd.to_datetime(data[date_col], errors='coerce')

# YearMonth column
data["YearMonth"] = data[date_col].dt.to_period("M").astype(str)

# monthly aggregation: count of TransactionID and sum of TransactionAmount
monthly = (data.groupby("YearMonth")
             .agg(tx_count=("TransactionID", "count"),
                  total_amount=("TransactionAmount", "sum"))
             .reset_index())

# convert YearMonth back to datetime for plotting (first day of month)
monthly['YearMonth'] = pd.to_datetime(monthly['YearMonth'])

# Plot 1: Monthly Transaction Count
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(monthly["YearMonth"], monthly["tx_count"], marker="o", linewidth=2)
ax.set_title("Monthly Transaction Count")
ax.set_xlabel("Transactions (Month)")
ax.set_ylabel("Transactions")
plt.xticks(rotation=45)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Plot 2: Monthly Total Amount (in millions)
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(monthly["YearMonth"], monthly["total_amount"] / 1e6, marker="o", linewidth=2)
ax.set_title("Monthly Total Amount")
ax.set_xlabel("Month")
ax.set_ylabel("Amount (Million INR)")
plt.xticks(rotation=45)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# display aggregated table
display(monthly)
```

## 7. RFM Segmentation

**Definition:**  
RFM stands for **Recency**, **Frequency**, and **Monetary**  three key behavioral metrics used to evaluate customer engagement and value.

### Components

- **Recency (R):** Number of days since the customers most recent transaction.  
  *Interpretation:* Lower values indicate more recent activity and higher engagement.

- **Frequency (F):** Total number of transactions completed by the customer.  
  *Interpretation:* Higher values reflect repeat purchases and stronger loyalty.

- **Monetary (M):** Total amount spent by the customer over the analysis period.  
  *Interpretation:* Higher spending signals greater customer value and profitability.

### Why it matters
RFM analysis helps classify customers into meaningful segments  such as **loyal**, **at-risk**, or **new**  enabling data-driven marketing and retention strategies.


```python
snapshot_date = data["TransactionDate"].max() + pd.Timedelta(days=1)

rfm = data.groupby("CustomerID").agg(
    recency_days=("TransactionDate", lambda x: (snapshot_date - x.max()).days),
    frequency=("TransactionID", "count"),
    monetary=("TransactionAmount", "sum")
).reset_index()

print(f"RFM table: {len(rfm):,} customers")
rfm.head(10)
```

```python
# Save RFM table to CSV
output_path = "../data/rfm_table.csv"
rfm.to_csv(output_path, index=False)

print(f" RFM table saved successfully as rfm_table.csv")
```

```python
# --- Part 1: Calculate RFM Values ---
if {"CustomerID", "TransactionDate", "TransactionID", "TransactionAmount"}.issubset(data.columns):
    # Ensure TransactionDate is in datetime format
    data["TransactionDate"] = pd.to_datetime(data["TransactionDate"])
    
    ref_date = data["TransactionDate"].max() + pd.Timedelta(days=1)
    
    rfm = (
        data.groupby("CustomerID")
        .agg(
            Recency=("TransactionDate", lambda x: (ref_date - x.max()).days),
            Frequency=("TransactionID", "nunique"),
            Monetary=("TransactionAmount", "sum")
        )
        .reset_index()
    )
    
    # --- Part 2: Handle potential ties or zero-variance columns ---
    def safe_qcut(series, q=5, labels=None):
        # If all values are the same, return middle score
        if series.nunique() == 1:
            return pd.Series([3]*len(series), index=series.index)
        else:
            return pd.qcut(series.rank(method="first"), q=q, labels=labels)

    # --- Part 3: Quantile-based Scoring ---
    rfm["R_Score"] = safe_qcut(rfm["Recency"], q=5, labels=[5, 4, 3, 2, 1])  # lower Recency = better
    rfm["F_Score"] = safe_qcut(rfm["Frequency"], q=5, labels=[1, 2, 3, 4, 5]) # higher better
    rfm["M_Score"] = safe_qcut(rfm["Monetary"], q=5, labels=[1, 2, 3, 4, 5])  # higher better
    
    # Convert to int and calculate total score
    rfm[["R_Score","F_Score","M_Score"]] = rfm[["R_Score","F_Score","M_Score"]].astype(int)
    rfm["RFM_Score"] = rfm[["R_Score", "F_Score", "M_Score"]].sum(axis=1)

    # --- Part 4: Dynamic Segment Definition ---
    # You can tweak these ranges or make them percentile based
    def segment(score):
        if score >= 13:
            return "Champions"
        elif score >= 10:
            return "Loyal"
        elif score >= 7:
            return "Potential Loyalist"
        elif score >= 5:
            return "At Risk"
        else:
            return "Lost"

    rfm["Segment"] = rfm["RFM_Score"].apply(segment)

    # --- Part 5: Display Results ---
    print(" RFM Segmentation Completed")
    display(rfm.head(10))
    
    print("\n Segment distribution:")
    display(rfm["Segment"].value_counts())
```
