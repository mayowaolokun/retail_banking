##### 2. Pasing Date

```python
# Parse and fix CustomerDOB; handling future date
def parse_customer_dob(dob_str):
    try:
        dt = pd.to_datetime(dob_str, dayfirst=True, errors='coerce')
        # Fix future DOB (e.g., 2094 -> 1994)
        if pd.notnull(dt) and dt > pd.Timestamp.now():
            dt = dt.replace(year=dt.year - 100)
        return dt
    except:
        return pd.NaT

# Parse and fix TransactionDate
def parse_transaction_date(date_str):
    try:
        dt = pd.to_datetime(date_str, dayfirst=True, errors='coerce')
        # Convert 16 -> 2016
        if pd.notnull(dt) and dt.year < 2000:
            dt = dt.replace(year=dt.year + 2000)
        return dt
    except:
        return pd.NaT

# Apply transformations ---
data['CustomerDOB'] = data['CustomerDOB'].apply(parse_customer_dob)
data['TransactionDate'] = data['TransactionDate'].apply(parse_transaction_date)

# --- Ensure dtype is datetime64
data['CustomerDOB'] = pd.to_datetime(data['CustomerDOB'], errors='coerce')
data['TransactionDate'] = pd.to_datetime(data['TransactionDate'], errors='coerce')

# Verify the datatypes
print(data[['CustomerDOB', 'TransactionDate']].dtypes)
print(data[['CustomerDOB', 'TransactionDate']].head(20))
```

##### 3. Combine Transaction date and Time for accurate analysis, such as hourly patterns.

```python
# Parse TransactionTime
def parse_transaction_time(t):
    """Convert HHMMSS numeric or string to a proper time object, handling invalid values safely."""
    if pd.isna(t):  # handle NaN
        return None
    t = ''.join(filter(str.isdigit, str(t)))  # remove any non-digit characters
    if len(t) != 6:  # if it's not exactly 6 digits, return None
        return None
    hour = int(t[0:2])
    minute = int(t[2:4])
    second = int(t[4:6])
    return time(hour, minute, second)

# Convert TransactionTime to time objects
data['TransactionTime'] = data['TransactionTime'].apply(parse_transaction_time)

# Combine TransactionDate and TransactionTime
data['TransactionDateTime'] = data.apply(
    lambda row: datetime.combine(row['TransactionDate'].date(), row['TransactionTime'])
                if pd.notnull(row['TransactionDate']) and row['TransactionTime'] is not None else pd.NaT,
    axis=1
)

# Convert to datetime64 (ensures correct dtype)
data['TransactionDateTime'] = pd.to_datetime(data['TransactionDateTime'], errors='coerce')

# Print the three columns ---
print(data[['TransactionDate', 'TransactionTime', 'TransactionDateTime']].head(20))
print(data[['TransactionDate', 'TransactionTime', 'TransactionDateTime']].dtypes)
```

#### 5. Outlier detection

##### Using IQR

```python
# Define a function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"--- {column} ---")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f}")
    print(f"Number of outliers in {column}: {len(outliers)}\n")
    return outliers

# Detect outliers in TransactionAmount and CustAccountBalance
outliers_transaction = detect_outliers_iqr(data, 'TransactionAmount')
outliers_balance = detect_outliers_iqr(data, 'CustAccountBalance')
```

##### Using Percentile

```python
def detect_outliers_percentile(df, column, lower_pct=1, upper_pct=99):
    lower_bound = np.percentile(df[column], lower_pct)
    upper_bound = np.percentile(df[column], upper_pct)

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"--- {column} ---")
    print(f"{lower_pct}th Percentile: {lower_bound:.2f}")
    print(f"{upper_pct}th Percentile: {upper_bound:.2f}")
    print(f"Number of outliers in {column}: {len(outliers)}\n")
    return outliers

# Detect outliers using percentile method
outliers_transaction = detect_outliers_percentile(data, 'TransactionAmount', lower_pct=1, upper_pct=99)
outliers_balance = detect_outliers_percentile(data, 'CustAccountBalance', lower_pct=1, upper_pct=99)
```

##### Creating Capped Copy for Visualization Only

```python
eda = data.copy()

# Columns to cap
cols_to_cap = ['TransactionAmount', 'CustAccountBalance']

for col in cols_to_cap:
    lower, upper = eda[col].quantile([0.01, 0.99])   # 1st and 99th percentile
    eda[col] = eda[col].clip(lower, upper)

print("Created capped copy 'eda' for visualizations")
print("Original 'data' remains unchanged for modeling")

# Check summary statistics after capping
print(eda[cols_to_cap].describe().T)
```

#### Gender vs Transaction Amount & Account Balance

```python
# Exclude gender categories with very few records
valid_genders = data['CustGender'].value_counts()
valid_genders = valid_genders[valid_genders > 100].index  # threshold can be adjusted

filtered_data = data[data['CustGender'].isin(valid_genders)]

# Group by gender and calculate average
gender_summary = filtered_data.groupby('CustGender')[['TransactionAmount', 'CustAccountBalance']].mean().reset_index()

# Plot
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='CustGender', y='TransactionAmount', data=gender_summary, palette='viridis')
plt.title('Average Transaction Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Transaction Amount (INR)')

plt.subplot(1, 2, 2)
sns.barplot(x='CustGender', y='CustAccountBalance', data=gender_summary, palette='magma')
plt.title('Average Account Balance by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Account Balance (INR)')

plt.tight_layout()
plt.show()
```

```python
#  Get top 10 locations by frequency
top_locations = data['CustLocation'].value_counts().head(10).index
subset = data[data['CustLocation'].isin(top_locations)]

plt.figure(figsize=(16, 6))
sns.barplot(
    x='CustLocation',
    y='TransactionAmount',
    data=subset,
    estimator='mean',           # average transaction per location
    errorbar=None,              # suppress CI bar (new seaborn syntax)
    hue='CustLocation',         # needed when using palette
    palette='viridis',
    legend=False
)
plt.title('Average Transaction Amount by Top 10 Locations')
plt.xlabel('Location')
plt.ylabel('Transaction Amount (INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
